```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load Data

```python
df2005 = pd.read_csv('desktop/mortality_selected/2005.csv')
df2006 = pd.read_csv('desktop/mortality_selected/2006.csv')
df2007 = pd.read_csv('desktop/mortality_selected/2007.csv')
df2008 = pd.read_csv('desktop/mortality_selected/2008.csv')
df2009 = pd.read_csv('desktop/mortality_selected/2009.csv')
df2010 = pd.read_csv('desktop/mortality_selected/2010.csv')
df2011 = pd.read_csv('desktop/mortality_selected/2011.csv')
df2012 = pd.read_csv('desktop/mortality_selected/2012.csv')
df2013 = pd.read_csv('desktop/mortality_selected/2013.csv')
df2014 = pd.read_csv('desktop/mortality_selected/2014.csv')
df2015 = pd.read_csv('desktop/mortality_selected/2015.csv')
df = pd.concat([df2005,df2006,df2007,df2008,df2009,df2010,df2011,df2012,df2013,df2014,df2015])
```

```python
a = df.groupby(['current_data_year','5y_age_band_recode','month_of_death','sex','marital_status',
               'race','rankable_cause_recode']).size().reset_index(name = 'size')
a
```

# cause 7

```python
cause_7 = a[a['rankable_cause_recode']==7]
cause_7
```

```python
print(np.unique(cause_7['size']))
binblock=np.arange(0, 1500, 30)
plt.hist(cause_7['size'], bins = binblock, rwidth=0.9)
plt.title("size")
plt.show()
```

```python
print(np.unique(cause_7['current_data_year']))
binblock=np.arange(2005, 2016, 1)
plt.hist(cause_7['current_data_year'], bins = binblock, rwidth=0.9)
plt.title("current_data_year")
plt.show()
```

```python
print(np.unique(cause_7['5y_age_band_recode']))
binblock=np.arange(0, 30, 1)
plt.hist(cause_7['5y_age_band_recode'], bins = binblock, rwidth=0.9)
plt.title("histogram")
plt.show()
```

```python
print(np.unique(cause_7['month_of_death']))
binblock=np.arange(0, 15, 1)
plt.hist(cause_7['month_of_death'], bins = binblock, rwidth=0.9)
plt.title("histogram")
plt.show()
```

```python
print(np.unique(cause_7['race']))
binblock=np.arange(0, 90, 1)
plt.hist(cause_7['race'], bins = binblock,rwidth=0.9)
plt.title("histogram")
plt.show()
```

```python
year_agg=cause_7[['size', 'current_data_year']].groupby('current_data_year').mean().reset_index()
plt.scatter(year_agg['current_data_year'], year_agg['size'])
plt.title('current_data_year')
plt.show()
```

```python
month_agg=cause_7[['size', 'month_of_death']].groupby('month_of_death').mean().reset_index()
plt.scatter(month_agg['month_of_death'], month_agg['size'])
plt.title('month_of_death')
plt.show()
```

```python
sex_agg=cause_7[['size', 'sex']].groupby('sex').mean().reset_index()
plt.scatter(sex_agg['sex'], sex_agg['size'])
plt.title('sex')
plt.show()
```

```python
race_agg=cause_7[['size', 'race']].groupby('race').mean().reset_index()
plt.scatter(race_agg['race'], race_agg['size'])
plt.title('race')
plt.show()
```

```python
y = cause_7['size']
xcolumns = ['current_data_year','5y_age_band_recode','month_of_death','race','sex','marital_status']
x = cause_7[xcolumns]
```

```python
age_1hot = pd.get_dummies(x['5y_age_band_recode'],prefix = 'age')
month_of_death_1hot = pd.get_dummies(x['month_of_death'],prefix = 'month_of_death')
sex_1hot = pd.get_dummies(x['sex'],prefix = 'sex')
marital_status_1hot = pd.get_dummies(x['marital_status'],prefix = 'marital_status')
race_1hot = pd.get_dummies(x['race'],prefix = 'race')
```

```python
X = pd.concat([x,age_1hot,month_of_death_1hot,sex_1hot,marital_status_1hot,race_1hot], axis=1)
X = X.drop('month_of_death', axis = 1)
X = X.drop('sex', axis = 1)
X = X.drop('marital_status', axis = 1)
X = X.drop('race', axis = 1)
X = X.drop('5y_age_band_recode', axis = 1)
X
```

```python
import statsmodels.api as sm
X=sm.add_constant(X)
cause_7_model = sm.GLM(y, X, family=sm.families.Poisson()).fit()
print (cause_7_model.summary())
```

```python
ypred=cause_7_model.predict(X)
cause_7['ypred']=ypred
```

```python
cause_7_model.aic
```

```python
cause_7_year_agg=cause_7[['ypred','size', 'current_data_year']].groupby('current_data_year').mean().reset_index()
cause_7_year_agg.plot(x='current_data_year', y=['size', 'ypred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```

```python
cause_7_age_agg=cause_7[['ypred','size', '5y_age_band_recode']].groupby('5y_age_band_recode').mean().reset_index()
cause_7_age_agg.plot(x='5y_age_band_recode', y=['size', 'ypred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```

```python
cause_7_month_agg=cause_7[['ypred','size', 'month_of_death']].groupby('month_of_death').mean().reset_index()
cause_7_month_agg.plot(x='month_of_death', y=['size', 'ypred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```

```python
cause_7_month_agg=cause_7[['ypred','size', 'sex']].groupby('sex').mean().reset_index()
cause_7_month_agg.plot(x='sex', y=['size', 'ypred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```

```python
cause_7_marital_status_agg=cause_7[['ypred','size', 'marital_status']].groupby('marital_status').mean().reset_index()
cause_7_marital_status_agg.plot(x='marital_status', y=['size', 'ypred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```
