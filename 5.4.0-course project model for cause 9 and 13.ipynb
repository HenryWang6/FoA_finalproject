```python
y = cause_13['size']
xcolumns = ['current_data_year','5y_age_band_recode','month_of_death','race','sex','marital_status']
x = cause_13[xcolumns]
```

```python
month_of_death_1hot = pd.get_dummies(x['month_of_death'],prefix = 'month_of_death')
sex_1hot = pd.get_dummies(x['sex'],prefix = 'sex')
marital_status_1hot = pd.get_dummies(x['marital_status'],prefix = 'marital_status')
race_1hot = pd.get_dummies(x['race'],prefix = 'race')
```

```python
X = pd.concat([x,month_of_death_1hot,sex_1hot,marital_status_1hot,race_1hot], axis=1)
X = X.drop('month_of_death', axis = 1)
X = X.drop('sex', axis = 1)
X = X.drop('marital_status', axis = 1)
X = X.drop('race', axis = 1)
X
```

```python
import statsmodels.api as sm
X=sm.add_constant(X)
cause_13_model = sm.GLM(y, X, family=sm.families.Poisson()).fit()
print (cause_13_model.summary())
```

```python
y_pred = cause_13_model.predict(X)
cause_13['y_pred'] = y_pred
cause_13
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y,y_pred))
```

```python
cause_13_year_agg=cause_13[['y_pred','size', 'current_data_year']].groupby('current_data_year').mean().reset_index()
cause_13_year_agg.plot(x='current_data_year', y=['size', 'y_pred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```

```python
cause_13_age_agg=cause_13[['y_pred','size', '5y_age_band_recode']].groupby('5y_age_band_recode').mean().reset_index()
cause_13_age_agg.plot(x='5y_age_band_recode', y=['size', 'y_pred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```

```python
cause_13_month_agg=cause_13[['y_pred','size', 'month_of_death']].groupby('month_of_death').mean().reset_index()
cause_13_month_agg.plot(x='month_of_death', y=['size', 'y_pred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```

```python
cause_13_sex_agg=cause_13[['y_pred','size', 'sex']].groupby('sex').mean().reset_index()
cause_13_sex_agg.plot(x='sex', y=['size', 'y_pred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```

```python
cause_13_marital_status_agg=cause_13[['y_pred','size', 'marital_status']].groupby('marital_status').mean().reset_index()
cause_13_marital_status_agg.plot(x='marital_status', y=['size', 'y_pred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```

```python
cause_13_race_agg=cause_13[['y_pred','size', 'race']].groupby('race').mean().reset_index()
cause_13_race_agg.plot(x='race', y=['size', 'y_pred'], figsize=(10,5), grid=True, style=['o','rx'])
plt.show()
```
